Iris flower classification in machine learning is a popular introductory classification problem. It involves training a model to categorize iris flowers into one of three species (setosa, versicolor, or virginica) based on their petal and sepal measurements. It's often used as a beginner's example to teach the fundamentals of supervised machine learning and classification algorithms.

---
INPUT TO THE MODEL :
1. Sepal length (in centimeters)
2. Sepal width (in centimeters)
3. Petal length (in centimeters)
4. Petal width (in centimeters)


OUTPUT FROM THE MODEL :

Predicted IrisFlower Species (setosa, versicolor, or virginica) based on the input features.

**IMPORTING LIBRARIES :**
1.   NumPy
2.   Pandas
3.   Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mplot
import seaborn as sns
%matplotlib inline
```



---



ðŸ“„ IMPORTING DATASET :

```python
columns =  ["ID","SepalLengthCm	","SepalWidthCm","PetalLengthCm	","PetalWidthCm","SPECIES"]

#LOADING DATA :
dataset=pd.read_csv('iris_dataset.csv',names=columns)
dataset.head()
```

```python
dataset.describe()
```



---



**VISUALIZATION OF DATASET**

Visualization of the dataset is important as it provides a clear and intuitive way to explore patterns, relationships, and outliers in the data, aiding in better understanding and making informed decisions in data analysis and modeling.

```python
sns.pairplot(dataset, hue='SPECIES')
```



---



**SEPERATING DATASET INPUT COLUMNS & OUTPUT COLUMN (SPECIES) :**

Separating dataset input columns (features) from the output column (target variable, e.g., species) is essential for preparing data for machine learning models, enabling the model to learn patterns and make predictions effectively.

```python
data = dataset.values

input = data[:,1:5]
output = data[:,5]

print(input)
print(output)
```



---



**SLITTING DATASET INTO TRAINING & TESTING :**

Splitting a dataset into training and testing sets is crucial to assess how well a machine learning model generalizes to new, unseen data, helping to evaluate its performance and avoid overfitting.

```python
from sklearn.model_selection import train_test_split
inputTRAIN, inputTEST, outputTRAIN, outputTEST = train_test_split(input, output, test_size=0.2)
print("Sr.No."+"\t"+"input TRAIN Data"+"\t"+"output TRAIN Data")
for i in range (len(inputTRAIN)):
  print(i+1,"\t",inputTRAIN[i],"\t",outputTRAIN[i])
print("-------------------------------------------------------")
print("\nSr.No."+"\t"+"input TEST Data"+"\t"+"output TEST Data")
for i in range (len(inputTEST)):
  print(i+1,"\t",inputTEST[i],"\t",outputTEST[i])
```



---



---



IMPORTING LIBRARY WHICH CALCULATE ACCURACY :

```python
from sklearn.metrics import accuracy_score
```



---



# **MODEL'S :**

**1) LOGISTIC REGRESSION :**

```python
# IMPORTING Logistic Regression :
from sklearn.linear_model import LogisticRegression

LR_Algorithm=LogisticRegression()

# TRAINING TO MODEL :
LR_Algorithm.fit(inputTRAIN,outputTRAIN)
```

PREDICTING THE RESULTS :

```python
LR_Prediction = LR_Algorithm.predict(inputTEST)
```

ACCURACY OF LogisticRegression :

```python
SCORE_1 = accuracy_score(outputTEST,LR_Prediction)*100
print(SCORE_1)
```

WHAT ARE ERRORS IN TESTING (IF PRESET) :

```python
for i in range (len(inputTEST)):
  if (LR_Prediction[i]!=outputTEST[i]):
    print(i+1," ERROR !!")
    print("Predicted Result : ",LR_Prediction[i])
    print("Actual Result : ",outputTEST[i])
```



---



**2) SUPPORT VECTOR CLASSIFIER :**

```python
# IMPORTING Support Vector Classifier (SVC) :
from sklearn.svm import SVC

SVC_Algorithm=SVC()
# TRAINING TO MODEL :
SVC_Algorithm.fit(inputTRAIN,outputTRAIN)
```

PREDICTING THE RESULTS :

```python
SVC_Prediction = SVC_Algorithm.predict(inputTEST)
```

ACCURACY OF SupportVectorClassifier :

```python
SCORE_2 = accuracy_score(outputTEST,SVC_Prediction)*100
print(SCORE_2)
```

WHAT ARE ERRORS IN TESTING (IF PRESET) :

```python
for i in range (len(inputTEST)):
  if (SVC_Prediction[i]!=outputTEST[i]):
    print(i+1," ERROR !!")
    print("Predicted Result : ",SVC_Prediction[i],"\n","Actual Result : ",outputTEST[i])
```



---



**3) DECISION TREE CLASSIFIER :**

```python
from sklearn.tree import DecisionTreeClassifier

DTC_Algorithm = DecisionTreeClassifier()

DTC_Algorithm.fit(inputTRAIN,outputTRAIN)
```

PREDICTING THE RESULTS :

```python
DTC_Prediction = DTC_Algorithm.predict(inputTEST)
```

ACCURACY OF DecisionTreeClassifier :

```python
SCORE_3 = accuracy_score(outputTEST,DTC_Prediction)*100
print(SCORE_3)
```

WHAT ARE ERRORS IN TESTING (IF PRESET) :

```python
for i in range (len(inputTEST)):
  if (DTC_Prediction[i]!=outputTEST[i]):
    print(i+1," ERROR !!")
    print("Predicted Result : ",DTC_Prediction[i],"\n","Actual Result : ",outputTEST[i])
```



---

```python
comparison = pd.DataFrame({'MODELS': ['LOGISTIC REGRESSION','SUPPORT VECTOR CLASSIFIER','DECISION TREE CLASSIFIER'],
                           'R2 SCORE': [SCORE_1, SCORE_2, SCORE_3]})
```
